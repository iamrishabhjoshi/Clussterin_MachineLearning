# Principal Component Analysis

```python
## To reduce the dimension of the data
```

```python
X=df.iloc[:,5:33]
from sklearn.decomposition import PCA
pca=PCA(n_components=2,random_state=42)
X_pca=pca.fit_transform(X)
```

```python
X_pca
# So now dimensions are reduced to Two
```

```python
# How much information we retain from the dataset
```

```python
pca.explained_variance_ratio_.cumsum()[1]
```

# Model Training

### K-Mean Clustering

```python
## Making elbow to find the best value of K
```

```python
from sklearn.cluster import KMeans
distortions=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,random_state=0)
    km.fit(X_pca)
    distortions.append(km.inertia_)
plt.plot(range(1,11),distortions,marker='o')
plt.xlabel("K-value")
plt.ylabel("distorsion")
```

```python
## As the best Value of K is 3 so train model on value of K=3
```

```python
model=KMeans(n_clusters=3,init='k-means++',n_init=10,max_iter=300,random_state=0)
model.fit(X_pca)
```

```python
y=model.predict(X_pca)
```

```python
y
```

```python
print(X_pca[y==0,0])
print(X_pca[y==0,1])
```

## Plot the Graph to show the Clusters

```python
plt.scatter(X_pca[y==0,0],X_pca[y==0,1],s=50,c="red",label="cluster 1")
plt.scatter(X_pca[y==1,0],X_pca[y==1,1],s=50,c="yellow",label="cluster 2")
plt.scatter(X_pca[y==2,0],X_pca[y==2,1],s=50,c="green",label="cluster 3")
plt.title("Cluster graph")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
```

## AgglomerativeClustering

```python
## Another method to find the no. of clusters
# Dendogram
import scipy.cluster.hierarchy as heir
dendogram=heir.dendrogram(heir.linkage(X_pca,method='ward'))
plt.title("Dendogram")
plt.xlabel("Questions")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
y=model.fit_predict(X_pca)
```

```python
plt.scatter(X_pca[y==0,0],X_pca[y==0,1],s=50,c="red",label="cluster 1")
plt.scatter(X_pca[y==1,0],X_pca[y==1,1],s=50,c="yellow",label="cluster 2")
plt.title("Cluster graph")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
```
